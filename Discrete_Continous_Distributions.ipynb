```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# CONTINOUS DISTRIBUTIONS
x_uniform = np.linspace(0, 1, 1000)
y_uniform = stats.uniform.pdf(x_uniform)

x_normal = np.linspace(-3, 3, 1000)
y_normal = stats.norm.pdf(x_normal)

x_exponential = np.linspace(0, 4, 1000)
y_exponential = stats.expon.pdf(x_exponential)

x_beta = np.linspace(0, 1, 1000)
y_beta = stats.beta.pdf(x_beta, 2, 5)

x_gamma = np.linspace(0, 10, 1000)
y_gamma = stats.gamma.pdf(x_gamma, 2)

x_lognormal = np.linspace(0, 3, 1000)
y_lognormal = stats.lognorm.pdf(x_lognormal, s=0.954, scale=np.exp(0))

x_chi2 = np.linspace(0, 8, 1000)
y_chi2 = stats.chi2.pdf(x_chi2, df=2)

x_t = np.linspace(-5, 5, 1000)
y_t = stats.t.pdf(x_t, df=5)

x_f = np.linspace(0.01, 5, 1000)
y_f = stats.f.pdf(x_f, 5, 2)

# DISCRETE DISTRIBUTIONS
x_binomial = np.arange(20)
y_binomial = stats.binom.pmf(x_binomial, n=20, p=0.5)

x_poisson = np.arange(20)
y_poisson = stats.poisson.pmf(x_poisson, mu=3)

x_geometric = np.arange(1, 22)
y_geometric = stats.geom.pmf(x_geometric, p=0.1)

x_hypergeom = np.arange(20)
y_hypergeom = stats.hypergeom.pmf(x_hypergeom, M=50, n=10, N=5)

x_negbin = np.arange(20)
y_negbin = stats.nbinom.pmf(x_negbin, n=5, p=0.5)



fig, axes = plt.subplots(5, 3, figsize=(15, 25))
axes = axes.flatten()
axes[0].plot(x_uniform, y_uniform)
axes[0].set_title('Uniform Distribution')
axes[1].plot(x_normal, y_normal)
axes[1].set_title('Normal Distribution')
axes[2].plot(x_exponential, y_exponential)
axes[2].set_title('Exponential Distribution')
axes[3].plot(x_beta, y_beta)
axes[3].set_title('Beta Distribution')
axes[4].plot(x_gamma, y_gamma)
axes[4].set_title('Gamma Distribution')
axes[5].plot(x_lognormal, y_lognormal)
axes[5].set_title('Log-Normal Distribution')
axes[6].plot(x_chi2, y_chi2)
axes[6].set_title('Chi-Squared Distribution')
axes[7].plot(x_t, y_t)
axes[7].set_title('Student\'s t-Distribution')
axes[8].plot(x_f, y_f)
axes[8].set_title('F Distribution')
axes[9].stem(x_binomial, y_binomial, use_line_collection=True)
axes[9].set_title('Binomial Distribution')
axes[10].stem(x_poisson, y_poisson, use_line_collection=True)
axes[10].set_title('Poisson Distribution'
axes[11].stem(x_geometric, y_geometric, use_line_collection=True)
axes[11].set_title('Geometric Distribution')
axes[12].stem(x_hypergeom, y_hypergeom, use_line_collection=True)
axes[12].set_title('Hypergeometric Distribution')
axes[13].stem(x_negbin, y_negbin, use_line_collection=True)
axes[13].set_title('Negative Binomial Distribution')

for ax in axes[14:]:
    ax.axis('off')
plt.tight_layout()
plt.show()
```

## Uniform Distribution Calculations:
**PDF (Probability Density Function):**
$$
f(x) = \frac{1}{b-a} \quad \text{for} \quad a \leq x \leq b
$$

**CDF (Cumulative Distribution Function):**
$$
F(x) = \frac{x - a}{b - a} \quad \text{for} \quad a \leq x \leq b
$$

## Normal Distribution Calculations:
**PDF:**
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left(\frac{x-\mu}{\sigma}\right)^2}
$$

**CDF:**
$$
F(x) = \frac{1}{2} \left[1 + \text{erf}\left(\frac{x - \mu}{\sigma \sqrt{2}}\right)\right]
$$

## Exponential Distribution Calculations:
**PDF (Probability Density Function):**
$$
f(x; \lambda) = \lambda e^{-\lambda x} \quad \text{for } x \geq 0
$$

**CDF (Cumulative Distribution Function):**
$$
F(x; \lambda) = 1 - e^{-\lambda x} \quad \text{for } x \geq 0
$$

## Poisson Distribution Calculations:
**PMF (Probability Mass Function) for \( k \) events in an interval:**
$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

**CDF for \( k \) events:**
$$
F(k; \lambda) = e^{-\lambda} \sum_{i=0}^{k} \frac{\lambda^i}{i!}
$$

```python
def dunif(x, min=0, max=1):
    # f(x) = 1/(b-a) for a ≤ x ≤ b
    return ((x >= min) & (x <= max)) / (max - min)

def punif(q, min=0, max=1):
    # F(x) = (x - a) / (b - a) for a ≤ x ≤ b
    return np.clip((q - min) / (max - min), 0, 1)

def rnorm(n, mean=0, sd=1):
    # Generates n samples from N(μ, σ^2)
    return np.random.normal(mean, sd, n)

def dnorm(x, mean=0, sd=1):
    # f(x) = (1 / (σ√2π)) * exp(-1/2((x-μ)/σ)^2)
    return (1 / (np.sqrt(2 * np.pi) * sd)) * np.exp(-0.5 * ((x - mean) / sd) ** 2)

def pnorm(q, mean=0, sd=1):
    # F(x) = 1/2[1 + erf((x - μ)/(σ√2))]
    return (1 + erf((q - mean) / (np.sqrt(2) * sd))) / 2

def qnorm(p, mean=0, sd=1):
    # Q(p) = μ + σ√2 * erfinv(2p - 1)
    return mean + np.sqrt(2) * sd * erfinv(2 * p - 1)

def qqnorm(x):
    # Returns sorted x and corresponding quantiles of a standard normal distribution
    return np.sort(x), np.sort(np.random.normal(size=len(x)))

def qqline(x):
    # Draws a reference line on a Q-Q plot
    sorted_x = np.sort(x)
    plt.plot(sorted_x, sorted_x, 'r--')

def dexp(x, rate=1):
    # f(x) = λ * exp(-λx) for x ≥ 0
    return rate * np.exp(-rate * x)

def pexp(q, rate=1):
    # F(x) = 1 - exp(-λx) for x ≥ 0
    return 1 - np.exp(-rate * q)

def rexp(n, rate=1):
    # Generates n samples from exponential distribution with rate λ
    return np.random.exponential(1 / rate, n)

def qexp(p, rate=1):
    # Q(p) = -ln(1-p)/λ
    return -np.log(1 - p) / rate

def plot(x, y):
    # Plot x vs y
    plt.plot(x, y)

def hist(x, bins=10):
    # Plots a histogram of x
    plt.hist(x, bins=bins)

def seq(start, end, step=1):
    # Generates a sequence from start to end with a given step
    return np.arange(start, end, step)

def lines(x, y):
    # Draws line segments connecting x and y
    plt.plot(x, y, '-')

def legend(labels):
    # Adds a legend to a plot
    plt.legend(labels)

def polygon(x, y):
    # Draws a filled polygon defined by x and y
    plt.fill(x, y)
```

```python
def to_standard_normal(x, mu, sigma):
    """ converting values from a normal distribution N(mu, sigma) to N(0,1) """
    return (x - mu) / sigma

def from_standard_normal(z, mu, sigma):
    """ converting values from N(0,1) to a normal distribution N(mu, sigma) """
    return z * sigma + mu

mu = 10
sigma = 2
data = np.random.normal(mu, sigma, 1000)
z_scores = to_standard_normal(data, mu, sigma)
reverted_data = from_standard_normal(z_scores, mu, sigma)

fig, ax = plt.subplots(2, 2, figsize=(12, 8))
ax[0, 0].hist(data, bins=30, alpha=0.7, color='blue')
ax[0, 0].set_title("Original Data: N(10, 2)")
ax[0, 1].hist(z_scores, bins=30, alpha=0.7, color='green')
ax[0, 1].set_title("Standard Normal Data: N(0, 1)")
ax[1, 0].hist(reverted_data, bins=30, alpha=0.7, color='red')
ax[1, 0].set_title("Reverted Data: N(10, 2)")
ax[1, 1].axis('off')
plt.tight_layout()
plt.show()
```
